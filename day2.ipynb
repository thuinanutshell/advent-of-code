```python
total_safe = 0
for report in number_lists:
    if report == sorted(report) or report == sorted(report, reverse=True):
        i, j = 0, 1
        sum_diffs = 0
        while j > i and j <= len(report)-1:
            diff = abs(report[i]-report[j])
            if diff >= 1 and diff <= 3:
                sum_diffs += 1
            i += 1
            j += 1
            if sum_diffs == len(report) - 1:
                total_safe += 1
```
